```python
# Bar chart for categorical data
ordered= ['Completed',  'FinalPaymentInProgress', 'Current','Defaulted','Chargedoff','Cancelled', 'Past Due (1-15 days)', 'Past Due (16-30 days)', 'Past Due (31-60 days)',
        'Past Due (61-90 days)', 'Past Due (91-120 days)',
       'Past Due (>120 days)']
plt.figure(figsize=(8,5))
base_color = sb.color_palette()[0]
# Return the Series having unique values
x = df['LoanStatus'].unique()
# Return the Series having frequency count of each unique value
y = df['LoanStatus'].value_counts(sort=False)
plt.xlabel('Loan current Status')
plt.ylabel('count')
plt.title("LoanStatus variable exploring")
# to make it readable
plt.xticks(rotation=80)
# To order it by the count
freq = df['LoanStatus'].value_counts()
# count_order = freq.index
# Plot the figure
sb.countplot(data=df, x='LoanStatus', color=base_color, order=ordered);
plt.show()
```

## Multivariate Exploration

> Now we think we should start using the most two variables of interst, income, and loan status that we have and start comparing them with other variables of interest like the dept to income ratio, home ownership and term

**We would like to start with the home ownership with the income to define if they are related or affect the loan status as we found above**

```python
def hue_graph(X, Y, Hue, log=True, title=''):
    fig = plt.figure(figsize = [8,6])
    ax = sb.pointplot(data = df, x =X , y = Y, hue = Hue,
               palette = 'Blues', linestyles = '', dodge = 0.4, order=ordered)
    plt.title(title)
    plt.ylabel('Monthly income in $')
    plt.xlabel('Loan status')
    if log == True:
        plt.yscale('log')
    ax.set_yticklabels([],minor = True)
    plt.xticks(rotation=80)
#     plt.xticks(ordered)
    plt.yticks([100, 1000, 10000, 100000], ['100', '1k', '10k', '100k'])
hue_graph('LoanStatus', 'StatedMonthlyIncome', 'IsBorrowerHomeowner', True, 'The relation between income, loan status and the effect of owning a home on them')
plt.show();
```

We find here that home ownership doesn't really have a strong corellation with the loan status, and the high income doesn't have a strong  realation that says they should have a house or the opposite though.


Then we should go for the TERM variable to see if it affects the loan status as well... 

```python
# use of functions
hue_graph('LoanStatus', 'StatedMonthlyIncome', 'Term', True, 'The effect of the payment Term on loans status and their relationship with income')
```

We find that high income poeple tend to pay in 12 months, the lowest period, and low income people tend to use the 60 month and 36 options.
